```python
# importing necessary libs to scrap the web

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

### Scraping

```python
url = "https://www.eapfoundation.com/vocab/academic/awllists/"

page=requests.get(url)
soup = bs(page.content, 'html.parser')

body_tag = soup.body

table = soup.find("table",{"class":"offset"})

table
```

```python
bolded_texts = table.find_all('b')
bolded_texts
```

```python
final_texts = []

for text in bolded_texts:
  final_texts.append(text.get_text())
```

```python
final_texts
```

### Little Preprocess

```python
import string
```

```python
for i, text in enumerate(final_texts):
  final_texts[i] = text.replace('\xa0','')
```

```python
final_texts
```

```python
final_text = ' '.join(str(e) for e in final_texts)
final_text
```
