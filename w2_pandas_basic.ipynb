### a. Impor request, BeautifulSoup, dan Pandas

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

### b. Mengakses website

```python
page=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YyFMmdJByV4")
soup = bs(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())
```

### c. Mengambil data hari dari website

```python
period_tags=seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
periods
```

### d. Menampilkan data tujuh hari dari website

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]

temps = [t.get_text() for t in
seven_day.select(".tombstone-container .temp")]

descs = [d['title'] for d in seven_day.select('.tombstone-container img')]

print(short_descs)
print(temps)
print(descs)
```

### e. Menampilkan tabel data tujuh hari

```python
# weather = pd.DataFrame.from_dict({
#     "period": periods,
#     "short_desc": short_descs,
#     "temp": temps,
#     "desc": descs,
# }, orient='index')
# weather = weather.transpose()
# weather


## Code diatas digunakan jika nilai dari tabel ada yang kosong

weather = pd.DataFrame({
  "period": periods,
  "short_desc": short_descs,
  "temp": temps,
  "desc": descs
})
weather
```

### f. Mengambil data temperatur dan diubah menjadi tipe int

```python
temp_nums = weather["temp"].str.extract("(?P<temp_num>\d+)", expand=False)
weather["temp_num"] = temp_nums.astype('int')
temp_nums
```

### g. Tampilkan rata-rata suhu cuaca

```python
total = 0
for x in temp_nums:
  total = total + int(x)

avg = total/temp_nums.size
avg
```

### h. Meilhat cuaca pada malam hari mengandung temperatur rendah

```python
is_night = weather['temp'].str.contains('Low')
weather['is_night'] = is_night
is_night
```

```python
# https://jogja.tribunnews.com/2022/09/14/4-rekomendasi-kuliner-bubur-ayam-di-jogja-cocok-buat-menu-sarapan

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url = "https://jogja.tribunnews.com/2022/09/14/4-rekomendasi-kuliner-bubur-ayam-di-jogja-cocok-buat-menu-sarapan?page=2"

page=requests.get(url)
soup = bs(page.content, 'html.parser')

head_tag = soup.head

body_tag = soup.body

judul = head_tag.contents[1].contents[0]

# print('Judul: ', judul)

tanggal = body_tag.time.contents[0]

# print('Tgl Penulisan: ', tanggal)

img_thumb = body_tag.find_all('img', class_='imgfull')[0]['src']

# print('Url Thumbnail: ', img_thumb)

strong_str = body_tag.find_all('strong')

bolded_text = []
for x in strong_str:
  bolded_text.append(x.text)

# print('Teks ter-bold: ', bolded_text)

paging = body_tag.find_all('div', {'class':'paging'})

pages = []
if(url[-3:]!='all'):
  pages = paging[0].find_all('a')

# print(pages)

pagination_link = []

for x in pages:
  pagination_link.append(x['href'])

# print(len(pagination_link))
# print(pagination_link)

num_of_pages = len(pagination_link)

embedded_links = body_tag.find_all('iframe')

list_of_links = []
for x in embedded_links:
  list_of_links.append(x['src'])

table = pd.DataFrame.from_dict({
  "Judul": judul,
  "Tanggal": tanggal,
  "Thumbnail": img_thumb,
  "Bolded-Tags": bolded_text,
  "Num-of-Pages": num_of_pages,
  "Embedded-links": list_of_links
}, orient='index')

# table = table.transpose() 

table.to_csv('tribunjogja-onepage.csv')

table
```

```python
## Scraping specific page (per-pagination)

def scrap_boldText(webpage, page):
  url = webpage + '?page=' + str(page)
  r = requests.get(url)

  soup = bs(r.content, 'html.parser')

  body_tag = soup.body

  strong_str = body_tag.find_all('strong')

  bolded_text = []
  for x in strong_str:
    bolded_text.append(x.text)

  return bolded_text

scrap_boldText("https://jogja.tribunnews.com/2022/09/14/4-rekomendasi-kuliner-bubur-ayam-di-jogja-cocok-buat-menu-sarapan", 'all')
```
